```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

from curvelops import FDCT2D
from curvelops.plot import create_colorbar
```

```python
m = 512
n = 512
x = np.zeros((m, n))
DCT = FDCT2D(x.shape)
```

```python
y = DCT * x

# Convert to a curvelet struct indexed by
# [scale, wedge (angle), x, y]
y_reshape = DCT.struct(y)
```

```python
s = 4
w = 0
a, b = y_reshape[s][w].shape
normalization = np.sqrt(y_reshape[s][w].size)
y_reshape[s][w][a // 2, b // 2] = 1 * normalization
y_reshape[s][w + len(y_reshape[s]) // 2][a // 2, b // 2] = -1j * normalization

y = DCT.vect(y_reshape)
```

```python
x = DCT.H @ y
```

```python
x_fk = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(x), norm="ortho"))
```

```python
vmin, vmax = 0.8 * np.array([-1, 1]) * np.abs(np.max(x))
fig, ax = plt.subplots(2, 2, figsize=(8, 8), sharex="row", sharey="row")

im = ax[0, 0].imshow(x.real.T, cmap="gray", vmin=vmin, vmax=vmax)
create_colorbar(im, ax[0, 0])

im = ax[0, 1].imshow(x.imag.T, cmap="gray", vmin=vmin, vmax=vmax)
create_colorbar(im, ax[0, 1])

im = ax[1, 0].imshow(np.abs(x_fk).T, cmap="turbo", vmin=0)
create_colorbar(im, ax[1, 0])

mask = np.abs(x_fk) > 0.01 * np.abs(x_fk).max()
im = ax[1, 1].imshow(
    (mask * np.angle(x_fk, deg=True)).T,
    cmap="twilight_shifted",
    vmin=-180,
    vmax=180,
)
cax, cb = create_colorbar(im, ax[1, 1])
cax.get_yaxis().set_major_locator(MultipleLocator(45))


ax[0, 0].set(
    xlim=(m // 2 - 50, m // 2 + 50),
    ylim=(n // 2 - 50, n // 2 + 50),
    title="Space domain (Real) magnified",
)
ax[0, 1].set(title="Space domain (Imag) magnified")
ax[1, 0].set(title="Frequency domain (Abs)")
ax[1, 1].set(title="Frequency domain (Phase)")
fig.tight_layout()
```

```python
def load_image_and_labels():
  import glob, re
  images = sorted(glob.glob('ADNI/**/*.nii', recursive=True))
  patients = dict(line.strip().split('\t') for line in open('data_groups.txt'))

  labels = []
  no_labels = []
  for image in images:
    subject_id = re.findall("ADNI_(\d{3}_S_\d{4})", image)[0]

    if subject_id not in patients:
      no_labels.append(image)
    else:
      labels.append(patients[subject_id])

  for image in no_labels:
    images.remove(image)

  assert len(images) == len(labels)

  return images, labels
```

# Generate Model Params

## Batched

```python
import numpy as np
from curvelops import FDCT3D
```

```python
ct3 = FDCT3D((182, 218, 182), axes=(-3, -2, -1), nbscales=3, nbangles_coarse=4, allcurvelets=False)
```

```python
structs = []

# loop through images
for image in images:

  # Load NIfTI image
  nifti_img = nib.load(image)

  # Get the image data
  img_data = nifti_img.get_fdata()

  ## generate curvelet coefficients
  cc = ct3 * img_data

  structs.append(ct3.struct(cc))
```

```python
scales = 3
wedges = []
for i in range(scales):
  wedges.append(len(structs[0][i]))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

multiscale_params = []
features = []

for i in range(scales):
  ith_scale_params = []

  for j in range(wedges[i]):
    phi_i_j = []
    for struct in structs:
      phi_i_j.append(struct[i][j].flatten())

    phi_i_j = np.array(phi_i_j).real

    # X: array-like of shape (n_samples, n_features)
    scaler = StandardScaler()
    phi_i_j = scaler.fit_transform(phi_i_j)
    # X: {array-like, sparse matrix} of shape (n_samples, n_features)
    pca = PCA(n_components=5)
    phi_i_j = pca.fit_transform(phi_i_j)

    features.append(phi_i_j)
    ith_scale_params.append((scaler, pca))

  multiscale_params.append(ith_scale_params)
```

```python
del structs
```

```python
features = np.hstack(features)
```

## Mini-batched

```python
import numpy as np
from curvelops import FDCT3D
```

```python
ct3 = FDCT3D((182, 218, 182), axes=(-3, -2, -1), nbscales=3, nbangles_coarse=4, allcurvelets=False)
```

```python
from sklearn.decomposition import IncrementalPCA
from sklearn.preprocessing import StandardScaler

item_count = len(images)
batch_size = 5
batches = item_count // batch_size

def initialize_params(scales, wedge_list):
  params = []

  for i in range(scales):
    wedge_cnt = wedge_list[i]
    scale_params = [(StandardScaler(), IncrementalPCA(n_components=5)) for _ in range(wedge_cnt)]
    params.append(scale_params)

  return params

wedge_list = []
batch_n = 0
structs = []

params = []
params_initialized = False

# loop through images
for image_n, image in enumerate(images):

  # Load NIfTI image
  nifti_img = nib.load(image)

  # Get the image data
  img_data = nifti_img.get_fdata()

  ## generate curvelet coefficients
  cc = ct3 * img_data

  cc = ct3.struct(cc)

  ## construct the structed version
  structs.append(cc)

  if not params_initialized:
    scales = 3
    for i in range(scales):
      wedge_list.append(len(structs[0][i]))
    params = initialize_params(scales, wedge_list)
    params_initialized = True

  if image_n % batch_size == 4:

    print(f'processing batch number {batch_n}')
    batch_n += 1

    # process
    for i in range(scales):
      for j in range(wedge_list[i]):
        phi_i_j = []
        for struct in structs:
          phi_i_j.append(struct[i][j].flatten())

        phi_i_j = np.array(phi_i_j).real

        (scaler, pca) = params[i][j]
        # X: array-like of shape (n_samples, n_features)
        scaler.partial_fit(phi_i_j)
        # X: {array-like, sparse matrix} of shape (n_samples, n_features)
        pca.partial_fit(phi_i_j)

    phi_i_j = []

    # reset structs
    structs = []

    # only to test the ram consumption
    # params = []
    # params_initialized = False

if len(structs) != 0:

  print(f'processing the final batch')

  # process
  for i in range(scales):
    for j in range(wedges[i]):
      phi_i_j = []
      for struct in structs:
        phi_i_j.append(struct[i][j].flatten())

      phi_i_j = np.array(phi_i_j).real

      (scaler, pca) = params[i][j]
      # X: array-like of shape (n_samples, n_features)
      scaler.partial_fit(phi_i_j)
      # X: {array-like, sparse matrix} of shape (n_samples, n_features)
      pca.partial_fit(phi_i_j)

  # reset structs
  structs = []

```

```python
features = []

# loop through images
for image in images:

  # Load NIfTI image
  nifti_img = nib.load(image)

  # Get the image data
  img_data = nifti_img.get_fdata()

  ## generate curvelet coefficients
  cc = ct3 * img_data

  cc = ct3.struct(cc)

  image_features = []

  for i, scale in enumerate(cc):
    for j, wedge in enumerate(scale):
      (scaler, pca) = params[i][j]
      norm = scaler.transform(np.array(wedge).reshape(1, -1).real)
      image_features.append(pca.transform(norm))

  features.append(np.hstack(image_features))

features = np.vstack(features)
```

# Parameter Search

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import roc_curve, precision_recall_curve, auc, make_scorer, recall_score, accuracy_score, precision_score, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
# param_grid = {
#     'min_samples_split': [3, 5, 10],
#     'n_estimators' : [100, 300],
#     'max_depth': [3, 5, 15, 25],
#     'max_features': [3, 5, 10, 20]
# }

param_grid = {
    'n_estimators': [50, 100, 200, 400],
    'max_depth': [3, 5, 10, 20],
    'min_samples_split': [4, 8, 16],
    'min_samples_leaf': [2, 4, 8],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# scorers = {
#     'precision_score': make_scorer(precision_score),
#     'recall_score': make_scorer(recall_score),
#     'accuracy_score': make_scorer(accuracy_score)
# }
scorers = make_scorer(accuracy_score)
```

```python
clf = RandomForestClassifier(n_jobs=-1, class_weight='balanced')
```

```python
def grid_search_wrapper(refit_score='precision_score'):
    """
    fits a GridSearchCV classifier using refit_score for optimization
    prints classifier performance metrics
    """
    skf = StratifiedKFold(n_splits=3)
    grid_search = GridSearchCV(clf, param_grid, scoring=scorers, refit=refit_score,
                           cv=skf, return_train_score=True, n_jobs=-1)
    grid_search.fit(X_train.values, y_train.values.ravel())

    # make the predictions
    y_pred = grid_search.predict(X_test.values)

    print('Best params for {}'.format(refit_score))
    print(grid_search.best_params_)

    # confusion matrix on the test data.
    print('\nConfusion matrix of Random Forest optimized for {} on the test data:'.format(refit_score))
    print(pd.DataFrame(confusion_matrix(y_test, y_pred),
                 columns=['pred_0', 'pred_1', 'pred_2'], index=['0', '1', '2']))
    return grid_search
```
